### Proposed Methods:

Implement pattern detection techniques to classify patterns in light data from stars. I would like to compare two methods for analyzing this light:

#### Approaches to finding exoplanets:
1) The transit detection method. This method looks at the level of brightness from a star and looks for dips in the brightness to detect planets that are passing directly between the star and earth. I would then find a model that can classify these types of patterns in light levels over time.
Some of the disadvantages of this method are that it can only detect planets that happen to have a plane of orbit that lines up exactly with our veiwpoint from Earth. This is probably a very small percentage of planets, however this method can be applied fairly easily to many stars. So this is the method that has actually lead to the discovery of most confirmed exoplanets. 
<br> The next method I would like to compare is the radial velocity detection method. This is a way to look at the wavelength of light coming from a star in an effort to find patterns of doppler shift that indicate a planet orbits the star. Some disadvantages of this method are that the data cannot be collected as easily, they often have to watch a small number of stars closely rather than being able to look at all the stars in the sky at once. The planet also has to have enough mass to create this "wobble" effect in the doppler shift. This method is often used to confirm planets discovered in other ways. 

#### Modeling methods to detect patterns:
Use either feature engineering to create a feature that describes the change in light over time. This could be a rolling average of the light levels. Or two rolling averages of different lag periods and if the cross over eachother at regular intervals, we would know that there is some pattern in the data.
There may also be a way to use clustering to discover these patterns as well.

#### Throw a model at it just for fun:


```python
non_target_cols = [col for col in df.columns if col != 'LABEL']
```

```python
X_train, X_holdout, y_train, y_holdout = train_test_split(df[non_target_cols], df['LABEL'])

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf.score(X_holdout, y_holdout)
```
